```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
from numpy import random
```

```python
#genearte random poisson with 2
x=random.poisson(lam=2,size=10)
print(x)
```

```python
#visualization with hist true and kde false
sns.distplot(random.poisson(lam=2,size=1000),hist=True,kde=False)
plt.show()
```

```python
#visualization with kde true and hist false
sns.distplot(random.poisson(lam=2,size=1000),hist=False,kde=True)
plt.show()
```

```python
#differenec between normal distribution and poisson distribution
#normal is continous whereas poisson is dicrete distribution
#how it may be nearly identical but for very large lam means equavalent to mean and for large std deviation
sns.distplot(random.normal(loc=50,scale=7,size=1000),hist=False,kde=True)
sns.distplot(random.poisson(lam=50,size=1000),hist=False,kde=True)
plt.show()
```

```python
#difference between poisson and binomail distribution
#a difference is subtle binomail is for discrete trials whereas poisson for continous trails
#they may be nearly similar but p should be nearly to 0 and for very large number of trails
sns.distplot(random.binomial(n=1000,p=0.01,size=1000),hist=False,kde=True)
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')
plt.show()
```
