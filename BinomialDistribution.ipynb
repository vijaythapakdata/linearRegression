```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
from numpy import random
```

```python
#binomial distribution
#(n= number pf trials,p= probability(for eg tossing coin 0.5),size=returned shape of the array)
x=random.binomial(n=10,p=0.5,size=10)
print(x)
```

```python
#visualization of binomial distribution
sns.distplot(random.binomial(n=50,p=0.5,size=1000),hist=False,kde=True,label="binomial")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#difference between normal and binomial distribution
sns.distplot(random.normal(loc=50,scale=5,size=1000),hist=False,kde=True,label="normal")
sns.distplot(random.binomial(n=100,p=0.5,size=1000),hist=False,kde=True,label="binomial")
plt.show()
```
