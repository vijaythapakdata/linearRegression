```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.distplot([0,3,5,4,6])
plt.show()
```

```python
import numpy as np
x=np.array([0,1,2,3,4,5])
sns.distplot(x)
plt.show()
```

```python
sns.distplot(x,hist=False,kde=True)
plt.show()
```

```python
y=np.array([6,7,8,9,10])
plt.xlabel("X-aixs")
plt.ylabel("Y-axis")
sns.distplot(y)
plt.show()
```
