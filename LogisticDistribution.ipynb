```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
%matplotlib inline
```

```python
#generate random logistic distribution array with mean 1 and std dev 2.0 size 2*3

x=random.logistic(loc=1,scale=2,size=(2,3))
print(x)
```

```python
#visualiztion
sns.distplot(random.logistic(size=1000),hist=False,kde=True)
plt.legend("Logistic distribution")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#differenec b/w normal distribution and logistic distribution
sns.distplot(random.normal(scale=2,size=1000),hist=False,kde=True,color="green")
sns.distplot(random.logistic(size=1000),hist=False,kde=True,color="red")
plt.legend(title="norma")
plt.legend("logistic")
plt.show()
```
